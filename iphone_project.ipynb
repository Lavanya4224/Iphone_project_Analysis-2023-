# **DATA CLEANING**

**Replacing MissingValues**

```python
df['Percentage_of_iPhone_Users'].fillna(df['Percentage_of_iPhone_Users'].median(), inplace=True)
df['No_of_iPhone_Sold_USA'].fillna(df['No_of_iPhone_Sold_USA'].mean(), inplace=True)
df['iOS_Market_Share'].fillna(df['iOS_Market_Share'].mean(), inplace=True)
df['Android_Market_Share'].fillna(df['Android_Market_Share'].mean(), inplace=True)
```

**Observation:**
The **null values have been successfully replaced** using the appropriate methods:

* **Median** was used for columns with outliers (e.g., *Percentage\_of\_iPhone\_Users*).
* **Mean** was used for data without outliers.

**Next Step:**
Before performing analysis, we need to **handle and modify the outliers** to avoid skewed results.

```python
from scipy.stats import zscore

numerical_columns = df.select_dtypes(include=[np.number]).columns
z_scores = np.abs(zscore(df[numerical_columns]))
outliers = (z_scores > 3).any(axis=1)  # Threshold of 3 standard deviations
print(outliers.sum())
```

**Observation:**
We used the **5-number summary** (Minimum, Q1, Median (Q2), Q3, Maximum) to **detect outliers** in the dataset.

Outliers were identified using the **IQR method** and then **replaced appropriately** to reduce their impact on the analysis, ensuring a cleaner and more reliable dataset.

```python
Q3=df['Percentage_of_iPhone_Users'].quantile(0.75)
Q1=df['Percentage_of_iPhone_Users'].quantile(0.25)
IQR=Q3-Q1
upperlimit=Q3+(1.5*IQR)
lowerlimit=Q1-(1.5*IQR)
df[(df['Percentage_of_iPhone_Users']>upperlimit)|(df['Percentage_of_iPhone_Users']<lowerlimit)]
```
